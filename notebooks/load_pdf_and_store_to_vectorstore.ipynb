```python
import pickle

def load_pickle(path):
    
    with open(file=path, mode="rb") as f:
        obj=pickle.load(f)

    print (f'Load from {path}')

def to_markdown(obj, path):

    with open(file=path, mode="w") as f:
        f.write(obj)

    print (f'To_Markdown: {path}')
```

```python
import os
from langchain_core.documents.base import Document
from langchain_community.document_loaders import UnstructuredMarkdownLoader

texts, tables = [], []
for i in range(7):
    markdown_path = f"backend/data/complex_pdf/pickle/parsed_llamaparse_{i}.md"

    loader = UnstructuredMarkdownLoader(
        file_path=markdown_path,
        mode="elements",
        chunking_strategy="by_title",
        strategy="hi_res",
        max_characters=4096,
        new_after_n_chars=4000,
        combine_text_under_n_chars=2000,
    )

    docs_llamaparse = loader.load()
    text_doc = Document(page_content=docs_llamaparse[0].page_content)
    text_doc.metadata['category'] = 'Text'
    texts.append(text_doc)
    if 'text_as_html' in docs_llamaparse[0].metadata.keys():
        table_doc = Document(page_content=docs_llamaparse[0].metadata['text_as_html'])
        table_doc.metadata['category'] = 'Table'
        tables.append(table_doc)
    # docs_table_llamaparse = [doc for doc in docs_llamaparse if doc.metadata["category"] == "Table"]
```

```python
from itertools import chain

docs_preprocessed = list(chain(texts,table_preprocessed))
docs_preprocessed
```

```python
index_name = 'pdf-doc-index-test'
index_body = {
    'settings': {
        'analysis': {
            'analyzer': {
                'my_analyzer': {
                    'char_filter':['html_strip'],
                    'tokenizer': 'nori',
                    'filter': ['nori_filter'],
                    'type': 'custom'
                }
            },
            'tokenizer': {
                'nori': {
                    'decompound_mode': 'mixed',
                    'discard_punctuation': 'true',
                    'type': 'nori_tokenizer'
                }
            },
            "filter": {
                "nori_filter": {
                    "type": "nori_part_of_speech",
                    "stoptags": [
                        "J", "XSV", "E", "IC","MAJ","NNB",
                        "SP", "SSC", "SSO",
                        "SC","SE","XSN","XSV",
                        "UNA","NA","VCP","VSV",
                        "VX"
                    ]
                }
            }
        },
        'index': {
            'knn': True,
            'knn.space_type': 'cosinesimil'  # Example space type
        }
    },
    'mappings': {
        'properties': {
            'metadata': {
                'properties': {
                    'category': {'type':'text'},
                    'origin_table': {'type':'text'},
                }
            },
            'text': {
                'analyzer': 'my_analyzer',
                'search_analyzer': 'my_analyzer',
                'type': 'text'
            },
            'vector_field': {
                'type': 'knn_vector',
                'dimension': "768"
            }
        }
    }
}
```

```python
from opensearchpy import OpenSearch

client = OpenSearch(
    hosts=[{'host': 'localhost', 'port': 9200}],
    http_compress=True,
    http_auth=('','!'),
    use_ssl=True,
    verify_certs=False,
)

index_exists = client.indices.exists(index_name)

if index_exists:
    response = client.indices.delete(index_name)
    print(response)

response = client.indices.create(index_name, body=index_body)
index_info = client.indices.get(index=index_name)
print("Index is created")
print(index_info)
```

```python
client.close()
```

```python
from langchain.vectorstores import OpenSearchVectorSearch
from langchain_core.embeddings import FakeEmbeddings

vector_db = OpenSearchVectorSearch(
    index_name=index_name,
    opensearch_url='https://localhost:9200',
    embedding_function=FakeEmbeddings(size=768),
    http_auth=('','!'), # http_auth
    is_aoss=False,
    engine="faiss",
    space_type="l2",
    bulk_size=100000,
    timeout=60,
    use_ssl=True,
    verify_certs=False,
    ssl_show_warn=False
)

vector_db.client.info()
```

```python
vector_db.add_documents(
    documents=docs_preprocessed,
    vector_field="vector_field",
    bulk_size=1000000,
)
```
